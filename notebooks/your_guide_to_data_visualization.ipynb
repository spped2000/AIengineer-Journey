# Data Visualization Best Practices

1. **Define a Clear Purpose**:
    - Identify Your Audience: Tailor visuals to their knowledge level and needs.
    - Set Objectives: Ensure the visualization answers specific questions or highlights key insights.
    - Choose the Right Visualization Type.

2. **Match Data to Visuals**:
    - Use appropriate chart types (e.g., line charts for trends, bar charts for comparisons).
    - Avoid misleading representations and ensure accuracy.
    - Simplify and focus.

3. **Eliminate Clutter**:
    - Remove unnecessary elements like excessive gridlines or redundant labels.
    - Highlight key information using color or size.
    - Use color thoughtfully.

4. **Maintain Consistency**:
    - Apply a consistent color scheme.
    - Ensure accessibility with distinguishable color palettes and sufficient contrast.
    - Provide context.

5. **Add Descriptive Titles and Labels**:
    - Clearly label axes, data points, and include units of measurement.
    - Include annotations to clarify complex or critical data aspects.

6. **Represent Data Honestly**:
    - Avoid manipulating scales or omitting data.
    - Use appropriate scales that reflect the data's range and distribution.

7. **Make It Interactive (When Applicable)**:
    - Enhance engagement with interactive elements like tooltips, filters, or zoom functions.
    - Ensure interactivity adds value and doesn't overwhelm or confuse the user.

8. **Test and Iterate**:
    - Seek feedback from peers or stakeholders on clarity and effectiveness.
    - Refine the visualization based on feedback and new insights.

# 1. Comparison
#### **Purpose**: Compare values across categories.

## Bar Chart

#### üìä Good For:
- Comparing different categories
- Visualizing discrete data
- Showing changes over time

#### üîç Important to Know:
- Bars can be vertical or horizontal
- Length proportional to value

#### ‚ö†Ô∏è Cautions/Awareness:
- Too many categories can clutter
- Consistent intervals between bars
- Avoid inconsistent scales

```python
# Import standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

# Load datasets
iris = pd.read_csv("/kaggle/input/iris/Iris.csv")
creditcard = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
titanic = pd.read_csv("/kaggle/input/titanic/train.csv")
housing = pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv")
netflix = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")
happiness = pd.read_csv('/kaggle/input/world-happiness/2019.csv')
covid = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
gapminder = pd.read_csv("/kaggle/input/gapminder-datacamp-2007/gapminder - gapminder.csv", index_col="Unnamed: 0")
```

```python
from matplotlib.ticker import PercentFormatter

# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=titanic, x="Sex", y="Survived", palette='viridis')

# Add title and labels
ax.set_title('Titanic Survival Rate by Sex', fontsize=16)
ax.set_xlabel('Sex', fontsize=14)
ax.set_ylabel('Survival Rate', fontsize=14)

# Customize the tick labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, fontsize=12)
ax.yaxis.set_major_formatter(PercentFormatter(1))

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```

## Count Plot

- Comparing counts of different groups

#### üîç Important to Know:
- Displays the count of observations in each category

#### ‚ö†Ô∏è Cautions/Awareness:

```python
# Create the count plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=titanic, x='Sex', palette=["#17becf", "#e377c2"])

# Add title and labels
ax.set_title('Distribution of Passengers by Gender on the Titanic', fontsize=16)
ax.set_xlabel('Sex', fontsize=14)
ax.set_ylabel('Count', fontsize=14)

# Customize the tick labels (increased fontsize)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, fontsize=12)

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```

## Grouped Count Plot

#### üìä Good For:

#### üîç Important to Know:

## Bullet Chart

#### üìä Good For:
- Comparing a single measure against a target
- Displaying performance metrics
- Visualizing progress towards goals

#### üîç Important to Know:
- Contains a main measure, a comparative measure, and qualitative ranges
- Compact and information-dense representation

#### ‚ö†Ô∏è Cautions/Awareness:
- Can be difficult to interpret without proper labeling
- Ensure qualitative ranges are clearly defined
- May not be suitable for comparing multiple measures at once

```python
# Ensure valid numeric data
df = happiness.dropna(subset=["Score"])
df = df[df["Score"] > 0]

# Define multiple countries for comparison
countries = ["United States", "UK", "Sweden", "China", "South Korea", "Australia", "Peru", "Venezuela", "South Africa"]

# Compute global benchmark values
global_avg = df["Score"].mean()
high_performance = df["Score"].quantile(0.75)
low_performance = df["Score"].quantile(0.25)

# Ensure valid ranges
score_min = df["Score"].min()
score_max = df["Score"].max()

if low_performance == high_performance:
    low_performance = score_min
    high_performance = score_max

# Plot setup
fig, ax = plt.subplots(figsize=(10, 6))
y_pos = np.arange(len(countries))

for i, country in enumerate(countries):
    selected = df[df["Country or region"] == country]
    happiness_score = selected["Score"].values[0] if not selected.empty else global_avg

    # Performance bands (Background bars)
    ax.barh(i, high_performance, color='lightgray', height=0.4, left=0, label="" if i == 0 else "")
    ax.barh(i, score_max - high_performance, color='gray', height=0.4, left=high_performance, label="Top Score" if i == 0 else "")

    # Actual happiness score (Measures)
    ax.barh(i, happiness_score, color='#007acc', height=0.25, left=0, label="Happiness Score" if i == 0 else "")

    # Global Average Marker
    ax.scatter(global_avg, i, color='red', s=30, label="Global Avg." if i == 0 else "")

# Labels and Aesthetics
ax.set_yticks(y_pos)
ax.set_yticklabels(countries)
ax.set_xlabel("Happiness Score")
ax.set_title("Happiness Scores of Selected Countries Compared to Global Performance")
ax.legend(loc="lower right")
ax.set_xlim([score_min, score_max])

plt.gca().invert_yaxis()  # Invert to match traditional bullet charts
plt.show()
```

# 2. Distribution
#### **Purpose**: Show data distribution to identify patterns or outliers. 

‚ùóThe next four plots (histogram, box plot, violin plot and swarm plot) are created using the same dataset to better visualize and compare the differences between these types of plots.

## Histogram

#### üìä Good For:
- Visualizing the distribution of continuous data
- Identifying the frequency of data within specified ranges (bins)
- Detecting patterns and outliers

#### üîç Important to Know:
- Represents data in bins or intervals
- Height of each bar indicates the number of observations within each bin

#### ‚ö†Ô∏è Cautions/Awareness:
- Choice of bin width can significantly affect interpretation
- Too few bins can oversimplify the data; too many can overcomplicate it
- Ensure the data range is covered appropriately

```python
# Convert infinite values to NaN if they exist and drop NaNs
housing['SalePrice'] = housing['SalePrice'].replace([np.inf, -np.inf], np.nan)
housing.dropna(subset=['SalePrice'], inplace=True)

# Create the histogram plot
plt.figure(figsize=(12, 7))
ax = sns.histplot(data=housing, x='SalePrice', kde=True, color='#e76f51', bins=30)

# Add title and labels
ax.set_title('Distribution of Sale Prices', fontsize=20)
ax.set_xlabel('Sale Price', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)

# Customize the tick labels
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# Show the plot
plt.show()
```

## Box Plot

#### üìä Good For:
- Visualizing the distribution of data through quartiles
- Identifying outliers
- Comparing distributions between groups

#### üîç Important to Know:
- Displays median, quartiles, and potential outliers
- Useful for summarizing the spread and skewness of data

#### ‚ö†Ô∏è Cautions/Awareness:
- Can be less informative for small datasets
- Requires understanding of statistical terms like quartiles and median
- Ensure the plot is not cluttered with too many categories

```python
# Create the boxplot
plt.figure(figsize=(12, 5))
ax = sns.boxplot(data=housing, x='SalePrice', color='#17becf')  

# Add title and labels
ax.set_title('Distribution of Sale Prices', fontsize=20)
ax.set_xlabel('Sale Price', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)

# Customize the tick labels
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# Show the plot
plt.show()
```

## Violin Plot

#### üìä Good For:
- Visualizing the distribution and density of data
- Comparing multiple distributions
- Identifying multimodal distributions

#### üîç Important to Know:
- Combines features of a box plot and a kernel density plot
- Displays the full distribution of the data

#### ‚ö†Ô∏è Cautions/Awareness:
- Requires a sufficient amount of data for accurate density estimation
- Can be complex to interpret without prior knowledge of density plots
- Ensure clear labeling and distinction between groups

```python
# Create the violin plot
plt.figure(figsize=(12, 7))
ax = sns.violinplot(data=housing, x='SalePrice', color='#d62728')  

# Add title and labels
ax.set_title('Distribution of Sale Prices', fontsize=20)
ax.set_xlabel('Sale Price', fontsize=16)
ax.set_ylabel('Density', fontsize=16)

# Customize the tick labels
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# Show the plot
plt.show()
```

## Swarm Plot

#### üìä Good For:
- Visualizing the distribution of data points within categories
- Showing individual data points and their distribution
- Avoiding overlap of data points

#### üîç Important to Know:
- Each point is plotted to avoid overlap, creating a non-overlapping scatter plot
- Useful for small to medium-sized datasets

#### ‚ö†Ô∏è Cautions/Awareness:
- Can become cluttered with large datasets
- Interpretation may be challenging with many categories
- Ensure points are clearly distinguishable and well-labeled

```python
# Create the swarm plot
plt.figure(figsize=(12, 7))
ax = sns.swarmplot(data=housing, x='SalePrice', color='#7b2cbf')  

# Add title and labels
ax.set_title('Distribution of Sale Prices', fontsize=20)
ax.set_xlabel('Sale Price', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)

# Customize the tick labels
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# Show the plot
plt.show()
```

# 3. Composition
#### **Purpose**: Show parts of a whole at a specific time.

## Pie Chart

#### üìä Good For:
- Showing the proportion of categories as parts of a whole
- Visualizing relative percentages
- Comparing parts of a dataset

#### üîç Important to Know:
- Each slice represents a category's proportion
- Best used with a limited number of categories

#### ‚ö†Ô∏è Cautions/Awareness:
- Difficult to interpret with too many slices
- Not suitable for precise comparisons
- Ensure labels and colors are distinct and clear

```python
# Calculate the distribution of species
species_distribution = iris['Species'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(species_distribution, labels=species_distribution.index, autopct='%1.1f%%', colors=sns.color_palette('colorblind'))
plt.title('Distribution of Iris Species', fontsize=16)
plt.show()
```

## Stacked Bar Chart

#### üìä Good For:
- Comparing overall and individual category contributions
- Visualizing cumulative data across categories
- Showing part-to-whole relationships over time

#### üîç Important to Know:
- Bars are divided into sub-bars representing different categories
- Each bar represents the total value of all sub-bars

#### ‚ö†Ô∏è Cautions/Awareness:
- Can become cluttered with too many categories
- Difficult to compare individual sub-bar lengths
- Ensure consistent color coding and clear labels

```python
# Define bins with rounded numbers
bin_edges = [0, 50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000]
housing['SalePriceBin'] = pd.cut(housing['SalePrice'], bins=bin_edges, labels=['0-50k', '50-100k', '100-150k', '150-200k', '200-250k', '250-300k', '300-350k', '350-400k'])

# Calculate the distribution of overall quality within each sale price bin
quality_distribution = housing.groupby(['SalePriceBin', 'OverallQual']).size().unstack()

# Medium-Light Playful Colors 
colors = ["#FF8866", "#C97DBB", "#FF96A8",   
          "#72B7E0", "#4593C3", "#F7C948", 
          "#5EC480", "#E3A370", "#B88ACB", "#F4A259"] 

# Create a stacked bar chart
quality_distribution.plot(kind='bar', stacked=True, color=colors, figsize=(12, 8))

# Add title and labels
plt.title('Distribution of Overall Quality Across Sale Price Bins')
plt.xlabel('Sale Price Bins')
plt.ylabel('Count')
plt.legend(title='Overall Quality', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Treemap

#### üìä Good For:
- Visualizing hierarchical data
- Showing proportions within a hierarchy
- Comparing the size of different categories

#### üîç Important to Know:
- Each rectangle represents a category, and its size is proportional to its value
- Nested rectangles show sub-categories

#### ‚ö†Ô∏è Cautions/Awareness:
- Can become cluttered with too many levels of hierarchy
- Small rectangles may be hard to interpret
- Ensure clear labeling and color differentiation

```python
import squarify

#  Bright & Playful colors
colors = ["#FF6F61", "#F7CAC9", "#92A8D1", "#034F84",
          "#D65076", "#EFC050", "#87A96B", "#B565A7", "#955251",
          "#009B77", "#DD4124", "#45B8AC", "#E15D44", "#5B5EA6"]

# Calculate frequencies
value_counts = netflix['country'].value_counts().iloc[:14]  # Show only the top 14 for readability
sizes = value_counts.values
labels = value_counts.index

# Create a treemap
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.8, color=colors)
plt.title('Distribution of Netflix Shows by Country')
plt.axis('off')
plt.show()
```

# 4. Trend
#### **Purpose**: Display data changes over time.

## Line Chart

#### üìä Good For:
- Visualizing trends over time
- Showing changes in data at equal intervals
- Comparing multiple data series

#### üîç Important to Know:
- Each point on the line represents a data value at a specific time
- Useful for continuous data

#### ‚ö†Ô∏è Cautions/Awareness:
- Too many lines can clutter the chart
- Ensure time intervals are consistent
- Be cautious of over-plotting data points

```python
import matplotlib.dates as mdates

# Select rows for US
covid_us = covid[covid['Country/Region'] == 'US']

# Set the color palette
sns.set_palette("Set2")

# Create the figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Create a line plot
sns.lineplot(x=pd.to_datetime(covid_us['ObservationDate']), y=covid_us['Confirmed'], linewidth=2, ax=ax)

# Enhance the chart
ax.set_title('Cumulative Number of Confirmed Cases in US', fontsize=20)
ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('Number of Confirmed Cases', fontsize=16)

# Improve x-axis date formatting
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=11)

# Add grid lines
ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

## Area Chart

#### üìä Good For:
- Visualizing cumulative data over time
- Showing trends and changes in data
- Comparing multiple data series with filled areas

#### üîç Important to Know:
- Similar to a line chart, but areas under the line are filled
- Useful for emphasizing the magnitude of changes over time

#### ‚ö†Ô∏è Cautions/Awareness:
- Can become cluttered with too many series
- Ensure areas are distinguishable with clear colors
- Be cautious of overlapping areas which can obscure data

```python
# Create the figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Create an area plot
ax.fill_between(pd.to_datetime(covid_us['ObservationDate']), covid_us['Confirmed'], color='skyblue', alpha=0.7)

# Enhance the chart
ax.set_title('Cumulative Number of Confirmed Cases in US', fontsize=20)
ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('Number of Confirmed Cases', fontsize=16)

# Improve x-axis date formatting
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=11)

# Add grid lines
ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

## Sparkline

#### üìä Good For:
- Embedding within text or tables to show trends
- Visualizing data in a compact space
- Showing simple data trends or variations

#### üîç Important to Know:
- Small, simple charts without axes or labels
- Often used to show changes over time

#### ‚ö†Ô∏è Cautions/Awareness:
- Limited detail due to size
- Not suitable for precise comparisons
- Ensure context is provided for interpretation

```python
# Filter data for three countries
countries = ['US', 'Canada', 'Mexico']
covid_data = covid[covid['Country/Region'].isin(countries)]

# Set the color palette
sns.set_palette("viridis")

# Create a figure for the sparklines
fig, axes = plt.subplots(1, 3, figsize=(15, 3), sharey=True)

# Create a sparkline for each country
for i, country in enumerate(countries):
    country_data = covid_data[covid_data['Country/Region'] == country]
    sns.lineplot(data=country_data, x=country_data.index, y='Confirmed', ax=axes[i], linewidth=1.5)
    axes[i].set_title(country, fontsize=14)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)
    axes[i].spines['left'].set_visible(False)
    axes[i].spines['bottom'].set_visible(False)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

# 5. Relationship
#### **Purpose**: Explore correlations between variables.

## Scatter Plot

#### üìä Good For:
- Visualizing the relationship between two continuous variables
- Identifying correlations, trends, and clusters
- Detecting outliers

#### üîç Important to Know:
- Each point represents an observation in the dataset
- Useful for large datasets with numerous observations

#### ‚ö†Ô∏è Cautions/Awareness:
- Overplotting can obscure data points
- Ensure points are clearly distinguishable
- Consider adding trend lines or regression lines for better interpretation.

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species', style='Species', palette='Set1', s=100)

# Labels and title
plt.xlabel("Sepal Length (cm)", fontsize=14)
plt.ylabel("Sepal Width (cm)", fontsize=14)
plt.title("Iris Dataset: Sepal Length vs. Sepal Width", fontsize=16)

# Show legend
plt.legend(title="Species")

# Show plot
plt.show()
```

## Bubble Chart

#### üìä Good For:
- Visualizing three dimensions of data
- Comparing entities based on size, x-axis, and y-axis values
- Identifying patterns and correlations

#### üîç Important to Know:
- Each bubble represents an entity with its position determined by the x and y values, and size determined by a third variable
- Useful for displaying relationships and magnitude

#### ‚ö†Ô∏è Cautions/Awareness:
- Too many bubbles can clutter the chart
- Ensure bubbles are distinguishable with clear labels and colors
- Be cautious of overlapping bubbles which can obscure data

```python
# Set style
sns.set_style("whitegrid")

# Create figure
plt.figure(figsize=(12, 7))

# Create bubble chart
sns.scatterplot(data=gapminder, 
                x="gdp_cap", 
                y="life_exp", 
                size="population", 
                hue="continent", 
                palette="Set1", 
                alpha=0.7, 
                sizes=(50, 1000))

# Scale x-axis to log scale for better visualization
plt.xscale("log")

# Labels and title
plt.xlabel("GDP per Capita (Log Scale)", fontsize=14)
plt.ylabel("Life Expectancy (Years)", fontsize=14)
plt.title("GDP per Capita vs. Life Expectancy (2007)", fontsize=16)

# Show legend
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=12, markerscale=0.5)

# Show plot
plt.show()
```

#### üìä Good For:
- Visualizing the intensity of data at different points
- Showing correlations between variables
- Identifying patterns and anomalies

#### üîç Important to Know:
- Represents data values with color gradients
- Useful for large datasets with complex relationships

#### ‚ö†Ô∏è Cautions/Awareness:

```python
# Calculate the correlation matrix
corr_matrix = titanic.corr(numeric_only=True)

# Create a heatmap
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, cbar_kws={"shrink": 0.8})

# Labels and title
plt.title('Correlation Heatmap of Titanic Dataset', fontsize=18)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
